```python
#improving the Header Names
data=data.rename(columns={'country':'Country','year':'Year','sex':'Gender','age':'Age','suicides_no':'SuicidesNo','population':'Population','suicides/100k pop':'Suicides100kPop','country-year':'CountryYear','HDI for year':'HDIForYear',' gdp_for_year ($) ':'GdpForYearMoney','gdp_per_capita ($)':'GdpPerCapitalMoney','generation':'Generation'})
```

```python
data.columns
```

```python
#Given that HDI only has incomplete data, 19456 out of 27820, it will be removed
#County Year is considered redundant to Country and year, and will be harder to process moving forward
```

```python
columns_to_drop = ['HDIForYear', 'CountryYear']

columns_exist = all(column in data.columns for column in columns_to_drop)

if columns_exist:
    data = data.drop(columns_to_drop, axis=1)
else:
    print("One or more columns to drop do not exist in the DataFrame.")
#columns have been dropped already
```

```python
# saving the csv after cleaning
data.to_csv('C:\\Users\\user\\Desktop\\DTDS Capstone\\cleaned_suicides_data.csv', index=False)
```

```python
#Showing the total no. of suicides per country
```

```python
suicides_per_country = data.groupby('Country')['SuicidesNo'].sum().reset_index()

suicides_per_country.columns = ['Country', 'Total Suicides']

suicides_per_country = suicides_per_country.sort_values(by='Total Suicides', ascending=False)

print(suicides_per_country)
```

```python
#creating a line graph for total yearly suicides per country
```

```python
suicides_data = data[['Country', 'Year', 'SuicidesNo']]

grouped_data = suicides_data.groupby(['Country', 'Year']).sum().reset_index()

plt.figure(figsize=(12, 8))

countries = grouped_data['Country'].unique()
for country in countries:
    country_data = grouped_data[grouped_data['Country'] == country]
    plt.plot(country_data['Year'], country_data['SuicidesNo'], label=country)

plt.xlabel('Year')
plt.ylabel('Number of Suicides')
plt.title('Number of Suicides per Country (1985-2016)')

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)

plt.grid(True)  # Add grid for better visualization

plt.show()
```
